```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker = "*", color = "green", linewidth = 2, label = "Boston")

plt.legend("Boston")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

#add axis labels
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Add the legend.
plt.legend()
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
#create the plot using plt.show
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
#create a horizontal bar chart
plt.barh(x_axis, y_axis, color='magenta', label='Boston')

plt.legend('Boston')

#axis label
plt.ylabel('Date')
plt.xlabel('Fare($)')

#invert y axis
plt.gca().invert_yaxis()

#add title
plt.title("PyBer Fare by Month")

#add a legend
plt.legend()
```

```python
#create plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color = "cyan")

#title
plt.title("PyBer Fare by Month")

#axis label
plt.ylabel('Date')
plt.xlabel('Fare($)')

#invert y axis
plt.gca().invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
#create scatter plot
plt.scatter(y_axis,x_axis, color = 'red')

#add legend
plt.legend('Chicago')

#add titles
plt.title("PyBer Fare by Month")

#axis label
plt.xlabel('Date')
plt.ylabel('Fare($)')

#invert y axis
plt.gca().invert_yaxis()
```

```python
#iterate through y-axis and multiply the point size by three to make them bigger
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

#create bubble chart
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#simplified way to loop through and increase point size
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Create Object-Oriented plot
fid, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color = "skyblue", edgecolor = "black", alpha=0.8, linewidth = 2)

#add legend
ax.legend(["Boston"])

#add title
plt.title("PyBer Fare by Month")

#axis label
plt.ylabel('Date')
plt.xlabel('Fare($)')

#add lim on fare
plt.xlim(0, 45)

#invert y axis
plt.gca().invert_yaxis()
```

```python
#increase plot size
plt.subplots(figsize=(8, 8))

#explode July
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#create plot
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

#show plot
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#create a plot Object-Oriented
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["lightpink", "lightcoral", "palegreen", "mediumaquamarine", "powderblue", "cornflowerblue", "mediumpurple", "hotpink", "plum", "peachpuff", "papayawhip", "mistyrose"]
#create plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,labels=x_axis,shadow=True,autopct='%.1f%%',explode=explode_values,startangle=90,counterclock=False, colors=colors)

plt.show()
```
